```python
# Importing required libraries -

import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuring the notebook -

import warnings
warnings.simplefilter('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

pd.set_option('display.float_format', lambda x: '%.3f' % x)

sns.set_style('whitegrid')

%matplotlib inline
```

```python
# Loading the dataset -

filepath = '/Users/kumarpersonal/Downloads/Heart-Disease-Pred/Data/merge_data.csv'
df = pd.read_csv(filepath)
```

```python
df.columns
```

```python
df.sample(5)
```

```python
# Shape of the dataset -

print("No. of rows: {}".format(df.shape[0]))
print("No. of cols: {}".format(df.shape[1]))
```

```python
# Checking the data type -

df.info(null_counts=False)
```

```python
# Checking unique values and their count -

def dist_check(df, col_name):
  print("Unique values : ", df[col_name].unique())
  print("Value counts : ")
  print(df[col_name].value_counts())

col_lst = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope']

for col in col_lst: 
  print(col, " -")
  dist_check(df, col)
  print("\n")
```

```python
# Replacing the '?' with the most occuring value -

col_lst = ['trestbps', 'chol', 'cp', 'fbs', 'restecg', 'thalach', 'exang', 'slope']

for col in col_lst:
    df[col].replace('?', df[col].value_counts().idxmax(), inplace=True)
```

```python
# Type conversion -

df['trestbps'] = df['trestbps'].astype('float64')
df['chol'] = df['chol'].astype('float64')
df['fbs'] = df['fbs'].astype('float64')
df['restecg'] = df['restecg'].astype('float64')
df['thalach'] = df['thalach'].astype('float64')
df['exang'] = df['exang'].astype('float64')
df['slope'] = df['slope'].astype('float64')
```

```python
# Checking for null values -

df.isnull().sum() / len(df) * 100
```

```python
# Checking for duplicate rows -

dup_rows = df[df.duplicated()]
print("No. of duplicate rows: ", dup_rows.shape[0])
```

```python
# Dropping duplicate rows -

df.drop_duplicates(inplace=True)
print("No. of rows remaining: ", df.shape[0])
```

```python
# Statistical summary of the data -

df.describe(include='all').T
```

```python
# Checking the distribution of target variable -

df['target'].value_counts(normalize=True)*100
```

```python
# Plotting a pie chart -

x = df['target'].value_counts().values
plt.figure(figsize=(7, 6))
plt.pie(x, center=(0, 0), radius=1.5, labels=df['target'].unique(), 
        autopct='%1.1f%%', pctdistance=0.5)
plt.axis('equal')
plt.show()
```

```python
# Checking for outliers -

def dist_plot(data, col):
    fig = plt.figure(figsize=(15, 3))
    fig.subplots_adjust(right=1.5)
    
    plt.subplot(1, 3, 1)
    sns.distplot(data[col], kde=True, color='red')
    plt.title('Histogram')
    
    plt.subplot(1, 3, 2)
    stats.probplot(data[col], dist='norm', fit=True, plot=plt)
    plt.title('Q-Q Plot')

    plt.subplot(1, 3, 3)
    sns.boxplot(y=data[col])
    plt.title('Boxplot')
    
    plt.show()

for col in df.columns:
    dist_plot(df, col)
```

```python
# Plotting the correlation heatmap -

plt.figure(figsize = (12, 7))
corr = df.corr(method='spearman')
mask = np.triu(np.ones_like(corr, dtype=bool))
cormat = sns.heatmap(corr, mask=mask, annot=True, 
                     cmap='Spectral', linewidths=1, 
                     vmin=-1, vmax=1, fmt=".2f")
cormat.set_title('Correlation Heatmap')
plt.show()
```

```python
# Checking the correlation with target variable -

plt.figure(figsize=(8, 5))
x = df.drop(columns=['target'], axis=1)
x.corrwith(df['target']).plot(kind='barh')
plt.show()
```

```python
# Creating and plotting age groups -

df['age_group'] = pd.cut(x=df['age'], 
                         bins=[df.age.min(), 40, 50, 60, df.age.max()], 
                         labels=['Grp1', 'Grp2', 'Grp3', 'Grp4'])

plt.figure(figsize=(10, 5))
ax1 = sns.countplot(x='age_group', hue='target', data=df)

for p in ax1.patches:
  ax1.annotate('{}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+1), ha='center')

plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(df, hue='target')
plt.show()
```

```python
df.drop(columns=['age_group'], 
        axis=1, inplace=True)

df.to_csv('/Users/kumarpersonal/Downloads/Heart-Disease-Pred/Data/final_data.csv', index=False)
```
